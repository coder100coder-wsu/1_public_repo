#### [24] groupby().sum() used
covid19 data is euphemism for covid19 deaths.

```python
# get sum of covid19 data by age_group
sum_df_grouped_covid19 = pd.DataFrame(df_3.groupby('AgeGroup')['COVID19'].agg('sum'))
```

#### [25] sort_values() used

```python
# sort values of sum(covid19), descending order
sum_df_grouped_covid19.sort_values(by='COVID19',ascending=False,inplace=True)
sum_df_grouped_covid19.reset_index(inplace=True, drop=False)
```

barplot; plot(kind='bar') used

```python
sum_df_grouped_covid19['COVID19'].plot(kind='bar', title="bar plot COVID19 vs age_group", 
                                       xlabel="age_group")\
        .set_xticklabels(rotation=90,labels=sum_df_grouped_covid19['AgeGroup'])
plt.ylabel('COVID19')
```

Per above plot, in general, despite severe overlap in age_group, covid19 data generally decreases with age_group. that is lower age tends to result in lower covid19 data. covid19 data is euphemism for covid19 deaths.

```python
# get sum of covid19 data by age_group, Sex
df_4 = pd.DataFrame(df_3.groupby(['AgeGroup','Sex'])['COVID19'].sum().reset_index())
df_4.sort_values(by=['COVID19'],ascending=True,inplace=True)
df_4.reset_index(inplace=True, drop=True)
print(df_4, "\n", type(df_4))
```

```python
df_4.boxplot('COVID19','AgeGroup',rot = 90,figsize=(10,6))
```

#### Explain Boxplot

Per the boxplot above, maximum COVID19 data seems to belong to age_group 50years and above. Due to severe overlap (e.g. 50-59 and 50-64) it is difficult to pin down other buckets of years.
age_group <18years, has the lowest COVID19 data.
